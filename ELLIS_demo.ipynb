```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'  # or any {'0', '1', '2'}
import tensorflow as tf
tf.get_logger().setLevel('ERROR')
!pip install trieste
!pip install tensorflow_probability==0.20.1
! git clone https://github.com/henrymoss/bodemo.git
from bodemo.scripts import run_bo
from trieste.acquisition.function import expected_improvement
import matplotlib.pyplot as plt
import trieste
from trieste.models.gpflow.builders import build_gpr
from trieste.models.gpflow import GaussianProcessRegression
```

```python
# Build a custom acquisition function
class custom_acquisition_function(expected_improvement):
    @tf.function
    def __call__(self, x ): # [N, d] -> [N, 1]
        mean, variance = self._model.predict(tf.squeeze(x, -2)) # [N, 1], [N, 1]
        
        # add own code here! Some examples are below

        # for expected improvement, uncomment the next two lines
        # normal = tfp.distributions.Normal(mean, tf.sqrt(variance))
        # acq_values =  (self._eta - mean) * normal.cdf(self._eta) + variance * normal.prob(self._eta)

        # lower confidence bound is the next two lines
        beta = 5.0 
        acq_values = -mean + beta * tf.sqrt(variance) # [N, 1]

        # perhaps try a different value for beta? 
        # perhaps a dynamic value that includes the current iteration counter?
        current_iter = tf.shape(self._model._model.data[0])[0]
        
        return acq_values
```

```python
def forrester(x):
    return ((6*x-2)**2)*tf.sin(12*x-4)

search_space = trieste.space.Box([0],[1]) # optimize over unit interval
initial_points = search_space.sample(3) # 3 initial points
data = trieste.data.Dataset(query_points=initial_points, observations=forrester(initial_points)) # evaluate the function at the initial points

for _ in range(1):
    model = GaussianProcessRegression(build_gpr(data, search_space), num_kernel_samples=0) # build_model
    model.optimize(data) # fit model to data
    acquisition = custom_acquisition_function(model, eta = tf.reduce_min(data.observations, axis=0)) # build acquisition function
    candidates = search_space.sample(100)[:,None,:]
    acq_values = acquisition(candidates) # evaluate acquisition function
    best = candidates[tf.argmax(acq_values[:,0])]

    # do some plotting
    plt.figure()
    x_plot = tf.linspace(0,1,100)[:,None]
    mean, var = model.predict(x_plot)
    plt.plot(x_plot, mean, color="blue", label="mean")
    plt.plot(x_plot, mean + 2*tf.sqrt(var), color="blue", alpha = 0.5, label="95% confidence interval")
    plt.plot(x_plot, mean - 2*tf.sqrt(var), color = "blue", alpha = 0.5)
    plt.plot(x_plot, acquisition(x_plot[:, None,:]), color="red", label="acquisition function")
    plt.scatter(data.query_points, data.observations, color="black", label="observations")
    plt.legend()
    data = data + trieste.data.Dataset(best, forrester(best)) # add new data point to dataset

```

```python
# Do BO on 2d Branin using expected improvement
log_regret_ei = run_bo(num_steps =10)
```

```python
# Do BO on 2d Branin using lower confidence bound (or own acq)
log_regret_lcb = run_bo(acquisition_fn = custom_acquisition_function, num_steps =10)
```

```python
#plot results 
plt.figure()
plt.xlabel("# Function Evals")
plt.ylabel("Log Regret")
plt.plot(log_regret_ei, color="blue", label="ei")
plt.plot(log_regret_lcb, color="black", label="lcb") 
plt.legend()
```
